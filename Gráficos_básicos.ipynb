```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
#Datos de ejemplo
x = np.linspace(0, 10, 100)
y = np.sin(x)
```

```python
#Gráfico de líneas
plt.figure(figsize=(10,4))
plt.plot(x, y)
plt.title('Gráfico de líneas')
plt.xlabel('x')
plt.ylabel('sin(x)')
plt.show()
```

```python
#Gráfico de Barras
categorias = ['A', 'B', 'C', 'D']
valores = [3, 7, 2, 5]
plt.figure(figsize=(10, 4))
plt.bar(categorias, valores)
plt.title('Gráfico de Barras')
plt.xlabel('Categorías')
plt.ylabel('Valores')
plt.show()
```

```python
#Gráfico de Dispersión
x = np.random.rand(50)
y = np.random.rand(50)

plt.figure(figsize=(10, 4))
plt.scatter(x, y)
plt.title('Gráfico de Dispersión')
plt.xlabel('x')
plt.ylabel('y')
plt.show()
```

```python
#Histograma
datos = np.random.randn(1000)

plt.figure(figsize=(10, 4))
plt.hist(datos, bins=30)
plt.title('Histograma')
plt.xlabel('Valores')
plt.ylabel('Frecuencia')
plt.show()
```
